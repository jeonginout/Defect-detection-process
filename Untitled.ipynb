```python
import pandas as pd
df = pd.read_csv("fab.csv")
df.head() #처음 5개 행 출력
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt
df_1 = df.set_index('SensorTime')
df_1 = df_1.sort_index(ascending = True)
df_1
```

```python
import numpy as np
ax = df_1.plot(y=['Sensor0', 'Sensor1', 'Sensor2'], figsize = (15, 6))
total_len = len(df_1.index)
ax.set_xticks(np.arange(0, total_len +1,100)) #x축에 몇 개만 씀
plt.xticks(rotation =45) #글자 기울임
plt.show()
```

```python
data = df_1.reset_index(drop=True)
data # SensorTime 지운 후 data 확인
```

```python
# 공정 합격, 불량 개수 count하고 pie chart로 확인 결과 pass 93.4%, fail 6.6%
count_pass = len(data.loc[data['Pass_Fail']==-1])
count_fail = len(data.loc[data['Pass_Fail']==1])

labels = ['Pass', 'Fail']
values = [count_pass, count_fail]
          
plt.pie(values, labels = labels, autopct = '%.1f')
plt.show()
```

```python
#null값 확인 > 많아서 안함
#data.info(verbose=True, null_counts=True)
```

```python
#결측치 확인
df.isnull().sum()

#결측치 비율 확인 > 딱히 안 해도
#missing_num = data.isnull().sum() / len(data)
#missing_num
```

```python
#결측치 평균 값으로 채우기
data = data.fillna(data.mean())

# 결측치 없어진 것을 알 수 있음
#missing_num = data.isnull().sum() / len(data)
#missing_num

data.isnull().sum()
```

```python
# Pass_Fail과 각 센서들 간의 상관계수의 절댓값 계산
data_a = data.corrwith(data['Pass_Fail']).abs() 

# Pass_Fail 열 삭제
data_Corr = data_a.drop(['Pass_Fail']) 
data_Corr
```

```python
# 상관성 그래프
ax = data_Corr.plot.bar(figsize = (15, 6), color = 'green')
total_len = len(data_Corr.index)
ax.set_xticks(np.arange(0, total_len + 1, 22))
plt.xlabel('Sensor')
plt.ylabel('Sum(pass fail_abs)')
plt.xticks(rotation = 60)
plt.show() 
```

```python
#내림차순 정리
data_Corr_sort = data_Corr.sort_values(ascending = False) 
data_Corr_sort
```

```python
# 상위 40개 확인 > 왜 굳이 40개인지 모르겠음. 다른 곳에서 더 차이가 많이 남
data_Corr_sort = data_Corr_sort[:40].reset_index() 
data_Corr_sort
```

```python
# 데이터에서 상관 계수가 높은 센서 40개 열만 추출하여
# data 변수에 저장 / 나머지 센서들은 없음
df = data[data_Corr_sort['index'].tolist()]
df
```

```python
df_x_data = df #pass_fail 없애고 상위 40개만 있는 데이터를 x
df_y_data = data['Pass_Fail'] #pass_fail만 있는 데이터를 y
```

```python
# train, test set 분할
#df_x_data, df_y_data로 각각 train 과 test 만들기 

from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df_x_data, df_y_data, train_size = 0.7, test_size = 0.3, shuffle = False)

# 센서들이 스케일이 달라서 정규화 필요함
ax = train_x.plot(y=['Sensor21', 'Sensor28', 'Sensor59'], figsize = (15,6)) 

test_y[20:30]
```

```python
#정규화

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train_minmax_scaled = scaler.fit_transform(train_x)

# test_x도 정규화 해주어야 함, test_x만을 가지고 정규화 하면 값이 달라져서 fit 안 넣음
test_x_scaled = scaler.transform(test_x)
```

```python
# 정규화된 train_set / 보기 편하게 데이터프레임으로 바꿈
train_x_scaled = pd.DataFrame(train_minmax_scaled, columns = train_x.columns, index = list(train_x.index.values))
train_x_scaled 

# 정규화된 test_set
test_x_scaled = pd.DataFrame(test_x_scaled, columns = test_x.columns, index = list(test_x.index.values))
test_x_scaled
```

```python
#정규화 된 것 확인(이전 그래프와 스케일 맞춰짐)
ax = train_x_scaled.plot(y = ['Sensor21', 'Sensor28', 'Sensor59'], figsize = (10, 6))
```

```python
#SMOTE로 데이터 보정
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=0)
x_train_sm, y_train_sm = SMOTE(random_state=0).fit_resample(train_x_scaled, train_y)

# -1을 0으로 바꾸기 : y_train_sm, test_y
y_train_sm_ch = y_train_sm.replace(-1, 0)
y_train_sm = y_train_sm_ch

test_y_ch = test_y.replace(-1, 0)
test_y = test_y_ch
```
